```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Create some toy 1D data ---
N = 1024                # number of points
L = 100.0               # domain length
x = np.linspace(0, L, N, endpoint=False)
print(x)

# Example signal: sum of sine waves + noise
f = (np.sin(2*np.pi*5*x/L))

# --- 2. Fourier transform ---
fhat = np.fft.fft(f)            # complex Fourier amplitudes
freqs = np.fft.fftfreq(N, d=L/N)  # corresponding k (1/units of x)

# --- 3. Power spectrum ---
P = np.abs(fhat)**2 / N          # raw power spectrum (normalized by N)

# --- 4. Keep only positive frequencies ---
mask = freqs > 0
k = freqs[mask]
Pk = P[mask]

# --- 5. Plot ---
plt.loglog(k, Pk, label="Power spectrum")
plt.xlabel("k (frequency)")
plt.ylabel("P(k)")
plt.legend()
plt.show()
```
